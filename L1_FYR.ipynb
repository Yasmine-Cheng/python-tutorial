### 等於
1.   = -> 宣告，將變數定義
2.   == -> 等於(值相同) 
3.   != -> 不等於(值不相同)，驚嘆號於程式中代表“不” 

```python
# 將x變數設定為數字3
# print()->將結果印出的功能function
x = 3
print('x是 ',x)
print('x的資料型態是 ',type(x))
```

```python
# (剛剛宣告過的x變數 是否為數字3，如果一樣結果為True，不同結果則為False)
print('x是3嗎？ ',x==3)
print('x是4嗎？ ',x==4)
```

```python
# (剛剛宣告過的x變數 是否不是數字3，如果不是數字4結果就為True)
print('x不是4嗎？ ',x!=4)
```

### 程式碼錯誤資訊
\(只列幾個常見錯誤說明及舉例)\
若畫面有紅色的內容彈出，顯示xxxError，則表示撰寫的程式碼中有錯誤的內容，導致無法運行。
通常將錯誤訊息複製下來放到google上後，都可以找到其他人是用甚麼方法解決的。

###### SyntaxError ：invalid syntax(語法錯誤，不正確的語法)
1. if, elif, else, for, while, class ,def 後面少了冒號
2. 等於符號用錯(==/=)
3. 單雙引號錯誤(需成雙成對出現)

```python
# for的後面需要加：，因此爆出錯誤
for i in range(3)
  print(i)
```

```python
# 雙等於才是一樣的意思，單等於是宣告的意思
y = 3
if y = 3:
  print('yes')
```

```python
# 一邊單引號，一邊雙引號->不對
print('yes")
```

###### IndentationError: unindent does not match any outer indentation level(縮排錯誤:未縮進與任何外部縮進級別都不匹配)


1.   縮排錯誤

```python
# 使用for-loop，設定程式跑3次，並將第幾次列印出，但for loop需要縮排，所以error(錯誤)寫說，expected(期待) an indented(縮排的) block(區段)
for i in range(3):
print(i)
```
