### 等於
1.   = -> 宣告，將變數定義
2.   == -> 等於(值相同) 
3.   != -> 不等於(值不相同)，驚嘆號於程式中代表“不” 

```python
# 將x變數設定為數字3
# print()->將結果印出的功能function
x = 3
print('x是 ',x)
print('x的資料型態是 ',type(x))
```

```python
# (剛剛宣告過的x變數 是否為數字3，如果一樣結果為True，不同結果則為False)
print('x是3嗎？ ',x==3)
print('x是4嗎？ ',x==4)
```

```python
# (剛剛宣告過的x變數 是否不是數字3，如果不是數字4結果就為True)
print('x不是4嗎？ ',x!=4)
```

### 程式的數字是從0開始算

```python
# 使用for-loop，設定程式跑3次，並將第幾次列印出
# range()，會產生序列的功能function
for i in range(3):
  print("現在是第 ",i ," 次")
```
